## Q2

In addition to the number of sales, the ice-cream shop also records its daily revenue ($ in sales). They’ have also captured the noon temperature for the last 10 days.

Create a scatter plot to represent the data. What can you conclude from the scatterplot? Make sure you have the appropriate labels for the x and y axis.


## Matplotlib

```python
Temp = [57.56, 61.52, 53.42, 59.36, 65.3, 71.78, 66.92, 77.18, 74.12, 64.58]
Sales = [215, 325, 185, 332, 406, 522, 412, 614, 544, 421]


plt.scatter(Temp, Sales )  
plt.xlabel('Temperature (°F)')
plt.ylabel('Sales')
plt.title('Sales at Noon')
```

## Seaborn

```python
sns.relplot(
    x =Temp,
    y =Sales).set(title = 'Sales at Noon', xlabel='Temperature (°F)', ylabel='Sales')
```

There seems to be a positive correlation of ice cream sales at noon. The hotter the temperature,more the ice creams are getting sold.

## Matplotlib

```python
df = pd.read_csv(r'C:\Users\bsila\Dexian_Bootcamp\code\DataPyLabExercise\Monthly Ice Cream Shop Sales by Flavor.csv')

fig, ax =  plt.subplots(layout = 'constrained')
ax.plot(df['Month'], df['Vanilla'], label = 'Vanilla')
ax.plot(df['Month'], df['Chocolate'], label = 'Chocolate')
ax.plot(df['Month'], df['Strawberry'], label = 'Strawberry')
ax.plot(df['Month'], df['Neapolitan'], label = 'Neapolitan')
ax.plot(df['Month'], df['Mango'], label = 'Mango')
ax.plot(df['Month'], df['Lime'], label = 'Lime')
ax.set_xlabel('Month')  
ax.set_ylabel('Units Sold')  
ax.set_title("Ice Cream Sales by Flavour per Month")
ax.legend()
```
