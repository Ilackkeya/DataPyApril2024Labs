## Correlation Trends

```python
sns.scatterplot( x = df['Units Sold'], y = df['Total Sales'])
plt.title('Total Sales by Units Sold')
plt.xlabel('Units Sold')
plt.ylabel('Total Sales ($)')
plt.show()
#Correlation coefficient
r = np.corrcoef(df['Units Sold'], df['Total Sales'])
r[0,1]
```

```python
sns.scatterplot( x = df['Price per Unit'], y = df['Operating Profit'])
plt.title('Operating Profit by Price Per Unit')
plt.xlabel('Price Per Unit')
plt.ylabel('Operating Profit ($)')
plt.show()
#Correlation coefficient
r = np.corrcoef(df['Price per Unit'], df['Operating Profit'])
r[0,1]
```

```python
sns.scatterplot( x = df['Units Sold'], y =  df['Operating Profit'])
plt.title('Operating Profit by Units Sold')
plt.xlabel('Units Sold')
plt.ylabel('Operating Profit ($)')
plt.show()
#Correlation coefficient
r = np.corrcoef(df['Units Sold'], df['Operating Profit'])
r[0,1]
```

```python
sns.scatterplot( x = df['Price per Unit'], y = df['Operating Margin']*100)
plt.title('Operating Margin by Price per Unit')
plt.xlabel('Price per Unit')
plt.ylabel('Operating Margin (%)')
plt.show()
#Correlation coefficient
r = np.corrcoef(df['Price per Unit'], df['Operating Margin'])
r[0,1]
```

```python
sns.scatterplot(x= df['Units Sold'], y = df['Operating Margin']*100)
plt.title('Operating Margin by Units Sold')
plt.xlabel('Units Sold')
plt.ylabel('Operating Margin (%)')
plt.show()
#Correlation coefficient
r = np.corrcoef(df['Units Sold'], df['Operating Margin'])
r[0,1]
```

```python
sns.scatterplot( x = df['Total Sales'], y = df['Operating Margin']*100)
plt.title('Operating Margin by Total Sales')
plt.xlabel('Total Sales')
plt.ylabel('Operating Margin (%)')
plt.show()
#Correlation coefficient
r = np.corrcoef(df['Total Sales'], df['Operating Margin'])
r[0,1]
```

```python
sns.scatterplot( x = df['Total Sales'], y = df['Operating Profit'])
plt.title('Operating Profit by Total Sales')
plt.xlabel('Total Sales')
plt.ylabel('Operating Profit ($)')
plt.show()
#Correlation coefficient
r = np.corrcoef(df['Total Sales'], df['Operating Profit'])
r[0,1]
```

The Operating profit, Operating margin shows weak positive and strong negative correlation with Price per unit. This is an area to be concentrated on increasing the profit and margin.
